```python
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
pwd = os.getcwd()
michelin1 = pd.read_csv(pwd + "/data_one-star-michelin-restaurants.csv")
michelin1_modified = michelin1.copy()
```

```python
michelin1_modified.dropna(inplace= True)
michelin1_modified
```

```python
michelin1_modified['price'].unique()
```

```python
conditions = [
   (michelin1_modified['price'] == '$'),
   (michelin1_modified['price'] == '$$'),
   (michelin1_modified['price'] == '$$$'),
   (michelin1_modified['price'] == '$$$$'),
   (michelin1_modified['price'] == '$$$$$')
]

results = (1,2,3,4,5)

michelin1_modified['price_range'] = np.select(conditions,results)
michelin1_modified
```

```python
condition = [
    (michelin1_modified['price_range'] == 1),
    (michelin1_modified['price_range'] == 2),
    (michelin1_modified['price_range'] == 3),
    (michelin1_modified['price_range'] == 4),
    (michelin1_modified['price_range'] == 5)
]

result = ('Inexpensive', 'Moderately Expensive','Expensive','Very Expensive','Extremely Expensive')

michelin1_modified['price_range_name'] = np.select(condition, result)

michelin1_modified
```

```python
michelin3 = pd.read_csv(pwd + '/data_three-stars-michelin-restaurants.csv')
michelin3_modified = michelin3.copy()
michelin3
```

```python
michelin3_modified.dropna(inplace= True)
michelin3_modified
```

```python
michelin3_modified['price'].unique()
```

```python
conditions3 = [
    (michelin3_modified['price'] == '$$'),
    (michelin3_modified['price'] == '$$$$'),
    (michelin3_modified['price'] == '$$$$$')
]

results3 = ('2','4','5') 

michelin3_modified['price_range'] = np.select(conditions3,results3)

michelin3_modified
```

```python
condition3 = [ 
    (michelin3_modified['price_range'] == '2'),
    (michelin3_modified['price_range'] == '4'),
    (michelin3_modified['price_range'] == '5')
]

result3 = ('Moderately Expensive','Very Expensive','Extremely Expensive')

michelin3_modified['price_range_name'] = np.select(condition3,result3)

michelin3_modified
```

```python
michelin3_modified.drop(columns=['price','url'], inplace= True)
michelin3_modified
```

```python
restaurants = pd.concat([michelin1_modified, michelin2_mofified, michelin3_modified])
restaurants
```

```python
star_pie = restaurants.groupby('stars').count().loc[:,'name']
plt.style.use('fivethirtyeight')

slices = star_pie
labels = ['1 Star', '2 Stars', '3 Stars']
colors = ['#008fd5', '#e5ae37', '#6d904f']

plt.pie(slices, labels= labels,colors= colors, wedgeprops={'edgecolor': 'black'})

plt.title('Percentage of Michelin Star Restaurants')
plt.tight_layout()
plt.show( )
```

```python
data_for_pie = restaurants.groupby('price_range_name').count().loc[:,'name']

plt.style.use('fivethirtyeight')

slices = data_for_pie
labels = ['Expensive', 'Extremely Expensive', 'Inexpensive', 'Moderately Expensive', 'Very Expensive']
colors = ['#008fd5', '#e5ae37', '#6d904f', 'purple', '#fc4f30']

plt.pie(slices, labels= labels,colors= colors, wedgeprops={'edgecolor': 'black'})

plt.title('Percentage of Restaurants by Price Range')
plt.tight_layout()
plt.show( )
```

```python
restaurants['price_range'] = pd.to_numeric(restaurants['price_range'])
cuisine_price_pt = restaurants.pivot_table(index='cuisine', values='price_range', aggfunc= np.mean).sort_values('price_range', ascending= 0).head(10)
cuisine_price_pt
```

```python
plt.xlabel('Price Range')
plt.ylabel('Cuisine')
plt.title('Michelin Star Cuisine Price Range')
plt.barh(cuisine_price_pt.index, cuisine_price_pt['price_range'], color='purple')
```

```python
plt.style.use('seaborn')

x = restaurants['latitude']
y = restaurants['longitude']

plt.scatter(x, y, cmap='summer',edgecolors= 'black', linewidths= 1, alpha= 0.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Michelin Restaurant Coordinates')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

plt.tight_layout()
```
